```python
import pandas as pd
import numpy as np

df = pd.read_csv("beverage_r.csv", sep=';')
```

## Задача 1

Разделить таблицу данных так, чтобы в обучающую выборку попало 66% наблюдений, а в тестовую выборку 33% наблюдений

```python
np.random.seed(42)
idx = np.random.permutation(len(df))

# вычисляем размер обучающей выборки
train_size = int(len(df) * 0.66)

# получаем индексы train и test
train_indices = idx[:train_size]
test_indices = idx[train_size:]

# формируем выборки
train_set = df.iloc[train_indices]
test_set = df.iloc[test_indices]

print("Train:", len(train_set))
print("Test:", len(test_set))
```

## Задача 2

Разделить таблицу данных на четыре одинаковых части.

```python
np.random.seed(42)

# Генерируем массив случайных чисел, равномерно распределенных от 0 до 1
r = np.random.uniform(0, 1, len(df))

# Вычисляем квантили для разделения данных на 4 равные части:
q1, q2, q3 = np.quantile(r, [0.25, 0.5, 0.75])

# Делим данные по квантилям
part1 = df.iloc[r <= q1, :]
part2 = df.iloc[(r > q1) & (r <= q2), :]
part3 = df.iloc[(r > q2) & (r <= q3), :]
part4 = df.iloc[r > q3, :]

print("Часть 1:", len(part1))
print("Часть 2:", len(part2))
print("Часть 3:", len(part3))
print("Часть 4:", len(part4))
```

## Задача 3

Разделить таблицу данных на две стратифицированные подвыборки.
По переменным кока-кола и минералка.

```python
df2 = df.copy()

np.random.seed(42)

df2['strata'] = df2['COKE'].astype(str) + '_' + df2['TAB'].astype(str)

train_idx = []
test_idx = []

train_ratio = 0.66

# стратифицированное разбиение
for stratum in df2['strata'].unique():
    stratum_indices = df2[df2['strata'] == stratum].index.values
    np.random.shuffle(stratum_indices)
    train_size = int(len(stratum_indices) * train_ratio)
    train_idx.extend(stratum_indices[:train_size])
    test_idx.extend(stratum_indices[train_size:])

train_set2 = df2.loc[train_idx].drop(columns=['strata'])
test_set2 = df2.loc[test_idx].drop(columns=['strata'])

print("Train shape:", train_set2.shape)
print("Test shape:", test_set2.shape)
print("Train mean:\n", train_set2[['COKE','TAB']].mean())
print("Test mean:\n", test_set2[['COKE','TAB']].mean())
```
